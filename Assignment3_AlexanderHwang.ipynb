```python
# This is just the csv in table form.
from datascience import *
import numpy as np
%matplotlib inline

imdb = Table.read_table('imdb.csv',encoding='cp1252')
imdb.show()
```

```python
# Here is the table sorted with the highest rated movie on top.
highestRating = imdb.select('Title', 'Rating')
highestRating = highestRating.sort('Rating', descending = True)
highestRating.show(3)

# The The Godfather and The Shawshank Redemption both share the #1 spot.
```

```python
# Here is the earliest movie in the database.
earliest = imdb.select('Title', 'Year')
earliest = earliest.sort('Year')
earliest.show(3)

# The Kid from 1921 is the earliest(oldest) movie in the database.
```

```python
# Here is the table sorted by rating and only showing the top 20 movies.
rating = imdb.sort('Rating', descending = True)
rating = rating.take[:20]
rating.show()
```

```python
# Here is the average rating of all the movies in the database.
ratingAverage = list(imdb.column('Rating'))
print("The average rating of all movies: " + str(np.average(ratingAverage)))

# The average rating of all movies is 8.2656
```

```python
# Here is the average rating for all movies form the 1940s.
sortedIMBD = imdb.sort('Year')
rating40s = sortedIMBD.where('Decade', are.equal_to(1940))
rating40sAverage = list(rating40s.column('Rating'))
print("The average rating of movies from the 40s: " + str(np.average(rating40sAverage)))

# The average rating of movies from the 40s is 8.257
```

```python
# Here is a table that contains all the movies that came out in 1999 on the database.
sortedIMBD = imdb.sort('Year')
ninety_nine = sortedIMBD.where('Year', are.equal_to(1999))
ninety_nine.show()
```

```python
# Here is a table with all movies with a rating higher than 8.5.
sortedIMBD = imdb.sort('Rating', descending = True)
really_highly_rated = sortedIMBD.where('Rating', are.above(8.5))
really_highly_rated.show()
```

```python
# Here are the ratings for 20th century and 21th century movies.
sortedIMBD = imdb.sort('Year')
century20 = sortedIMBD.where('Decade', are.below(2000))
century21 = sortedIMBD.where('Decade', are.above_or_equal_to(2000))

century20Average = list(century20.column('Rating'))
print("The average rating of movies before 2000 is: " + str(np.average(century20Average)))

century21Average = list(century21.column('Rating'))
print("The average rating of movies after 2000 is: " + str(np.average(century21Average)))

# The average rating of movies before 2000 is: 8.27836
# The average rating of movies after 2000 is 8.23797
```

```python
# Here is the proportion of movies from the 20th and 21th century.
sortedIMBD = imdb.sort('Year')
century20 = sortedIMBD.where('Decade', are.below(2000))
century21 = sortedIMBD.where('Decade', are.above_or_equal_to(2000))

num20Movies = century20.num_rows
num21Movies = century21.num_rows
numMovies = sortedIMBD.num_rows

print("There are: " + str(num20Movies) + 
      "\nmovies released in the 20th century form a total of: " + str(numMovies) +
      "\nmoves in the database." +
      "\nProportionally, about 2/3 of the movies were released in the 20th century.")

print("\nThere are: " + str(num21Movies) + 
      "\nmovies released in the 21th century form a total of: " + str(numMovies) +
      "\nmoves in the database." +
      "\nProportionally, about 1/3 of the movies were released in the 21th century.")

# 171/250 of the movies are from the 20th century, or about 2/3.
# 79/250 of the movies are from the 21th century, or about 1/3.
```
