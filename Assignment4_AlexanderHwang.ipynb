```python
# This is just the csv in table form.
import pandas as pd
olympics = pd.read_csv('olympics.csv')

# This option is used so I can force all rows to be displayed.
pd.set_option('display.max_rows', None)
```

```python
# First before we start let's look at the data.
olympics

# We notice that there are some issues with the column labels. 
# Some are missing, some make no sense, and others can be more clear.
```

```python
# I will replace basically all of the labels with ones that make more sense using .loc[].
olympics.loc[0, '0'] = 'Nation'
olympics.loc[0, '1'] = 'Summer Apperances'
olympics.loc[0, '2'] = 'Gold'
olympics.loc[0, '3'] = 'Silver'
olympics.loc[0, '4'] = 'Bronze'
olympics.loc[0, '5'] = 'Total'
olympics.loc[0, '6'] = 'Winter Apperances'
olympics.loc[0, '7'] = 'Gold'
olympics.loc[0, '8'] = 'Silver'
olympics.loc[0, '9'] = 'Bronze'
olympics.loc[0, '10'] = 'Total'
olympics.loc[0, '11'] = 'All Apperances'
olympics.loc[0, '12'] = 'Gold'
olympics.loc[0, '13'] = 'Silver'
olympics.loc[0, '14'] = 'Bronze'
olympics.loc[0, '15'] = 'Combined Total'

# Let's look at the changes.
olympics
```

```python
# Next let's look for duplicated data using .duplicated()
olympics.duplicated()
```

```python
# It looks like nothing is duplicated, but we can use drop_duplicates just for fun.
olympics.drop_duplicates(inplace = True)
```

```python
#Let's check if any rows or values are empty using .isnull().
olympics.isnull()
```

```python
# It looks like there are no empty rows or values, but we can drop rows and replace values for fun.
olympics.dropna(inplace = True)
olympics.fillna(0, inplace = True)
```

```python
# Let's look at the resulting data.
olympics
```
