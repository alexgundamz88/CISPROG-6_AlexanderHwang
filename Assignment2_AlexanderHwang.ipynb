```python
# This is just the csv in table form.
from datascience import *
import numpy as np
%matplotlib inline

pods = Table.read_table('podcasts.csv',encoding='cp1252')

# Release Date is a categorical value. When transfered directly from the csv the data type is string.
# Era is a categorical value. When transfered directly from the csv the data type is string. 
  # However, a coded value for 0,1,2(0=Ancient, 1=Early Modern, 2=Modern) should be used for easier manipulation of the data.
# Series is a categorical value. When transfered directly from the csv the data type is string. 
  # However, a bool should be used for eaiser manipulation of the data.
# Downloads is a numerical value. The data type should be int.

pods.show()
```

```python
# range_pods is a subtable that only contains rows from podcasts that been downloaded between 1,000,000 and 5,000,000 times.
# range_pods will be sorted in accending order of downloads for to read it easier.
range_pods = pods.where('Downloads', are.between(1000000, 5000000))
range_pods = range_pods.sort('Downloads')
range_pods.show()
```

```python
# ancient_pods_sorted is a subtable that only contains rows from podcasts that are from the ancient era, sorted by downloads.
ancient_pods_sorted = pods.where('Era', are.equal_to('Ancient'))
ancient_pods_sorted = ancient_pods_sorted.sort('Downloads', descending=True)

# I'll only show the top 5 more downloaded.
ancient_pods_sorted.show(5)

# ancient_date is a string that holds the date of the podcast with the most downloads about the ancient era.
ancient_date = ancient_pods_sorted.first('Release date')
print(ancient_date)
```
